```python
dfPurchaseOrderTotal = dfPurchaseOrder.merge(dfPurchaseOrderDetail, on = 'purchase_order_id', how = 'left').replace(np.NaN , 0)
# dfPurchaseOrderTotalNotComplete = dfPurchaseOrderTotal[(dfPurchaseOrderTotal['is_completed'] != 1) & (dfPurchaseOrderTotal['is_complete'] != 1)
#                                                        & (dfPurchaseOrderTotal['po_status'] == 'approval')]
dfPurchaseOrderDetail = dfPurchaseOrderTotal.merge(dfProductUnit, on = ['product_id' , 'product_unit_id'], how = 'left')
dfPurchaseOrderDetail['quantity_exchange'] = 0
# dfPurchaseOrderDetail[dfPurchaseOrderDetail['product_id'] == 6823]
```

```python
dfProductUnitExchange[dfProductUnitExchange['product_id'] == 6823]
```

```python
dfPurchaseOrderDetail1 = dfPurchaseOrderDetail.merge(dfProductUnitExchange , left_on = ['product_id','unit_id'] , right_on = ['product_id','result_unit_id'] , how = 'left')
dfPurchaseOrderDetail1['quantity_exchange_not_default'] = dfPurchaseOrderDetail1['purchase_order_quantity'] * dfPurchaseOrderDetail1['source_quantity'] / dfPurchaseOrderDetail1['result_quantity']
dfPurchaseOrderDetail1
```

```python
aMate = dfPurchaseOrderDetail1.loc[0].index
for i in range(0 , len(aMate)):
    if(aMate[i] == 'quantity_exchange'):
        a = i
    else: i = i + 1

for i in range(0 , dfPurchaseOrderDetail1.shape[0]):
    aMate = dfPurchaseOrderDetail1.loc[i]
    if(aMate['is_default'] == 1.0):
        dfPurchaseOrderDetail1.iloc[[aMate.name], [a]] = aMate['purchase_order_quantity']
    elif(aMate['is_default'] == 0.0):
        dfPurchaseOrderDetail1.iloc[[aMate.name], [a]] = aMate['quantity_exchange_not_default']

dfPurchaseOrderDetail1[(dfPurchaseOrderDetail1['is_default'] != 0) & (dfPurchaseOrderDetail1['is_default'] != 1)]
dfPurchaseOrderDetail1
```

```python
dfStockInPurchaseOrder
```

```python
dfPurchaseOrderDetailmerge = dfPurchaseOrderDetail1.merge(dfStockInPurchaseOrder, on = ['purchase_order_id' , 'product_id'], how = 'left').replace(np.NaN , 0)
dfPurchaseOrderDetailmerge['quantity_not_arrived'] = dfPurchaseOrderDetailmerge['quantity_exchange'] - dfPurchaseOrderDetailmerge['quantity_arrived']
dfPurchaseOrderDetailmerge
```

```python
dfPurchaseOrderDetailmerge.head(5)
```

```python
dfPurchaseOrderDetailmerge['raw_material_status'] = ''
aMate = dfPurchaseOrderDetailmerge.loc[0].index
for i in range(0 , len(aMate)):
    if(aMate[i] == 'raw_material_status'):
        a = i
    else: i = i + 1

for i in range(0 , len(dfPurchaseOrderDetailmerge)):
    aMate = dfPurchaseOrderDetailmerge.loc[i]
    if(aMate['is_completed'] == 1):
        dfPurchaseOrderDetailmerge.iloc[[aMate.name], [a]] = 'Hoàn tất'
    elif((aMate['is_completed'] == 0) & (aMate['is_complete'] == 0) & (aMate['quantity_arrived'] == 0)):
        dfPurchaseOrderDetailmerge.iloc[[aMate.name], [a]] = 'Chờ nhập kho'
    elif((aMate['is_completed'] == 0) & (aMate['is_complete'] == 0) & (aMate['quantity_arrived'] > 0) & (aMate['quantity_arrived'] < aMate['quantity_exchange'])):
        dfPurchaseOrderDetailmerge.iloc[[aMate.name], [a]] = 'Đã nhập kho 1 phần'
    elif((aMate['is_completed'] == 0) & (aMate['is_complete'] == 0) & (aMate['quantity_arrived'] > 0) & (aMate['quantity_arrived'] > aMate['quantity_exchange'])):
        dfPurchaseOrderDetailmerge.iloc[[aMate.name], [a]] = 'Nhập kho quá số lượng'
    elif((aMate['is_completed'] == 0) & (aMate['is_complete'] == 1)):
        dfPurchaseOrderDetailmerge.iloc[[aMate.name], [a]] = 'Đã nhập đủ'

dfPurchaseOrderDetailmerge
    
```

```python
dfPurchaseOrderDetailDA = dfPurchaseOrderDetailmerge[['code', 'name_po','total_amount' , 'purchase_order_id' , 'is_completed' , 'is_complete' , 'po_status', 'product_code' , 'supplier_code' , 'supplier_name' , 'purchase_order_quantity' , 'quantity_exchange' , 'quantity_arrived' , 'quantity_not_arrived'
                                                 , 'time_create' , 'estimated_delivery' , 'actual_delivery_time' , 'fullname' , 'raw_material_status']]

dfPurchaseOrderDetailDA.rename(columns = { 'code' : 'Mã ĐH' ,'name_po':'Tên ĐH' ,'total_amount':'Số lượng ĐH đặt' , 'purchase_order_id': 'ID đơn hàng', 'fullname' :'Người tạo ĐH', 'is_completed': 'ĐH hoàn tất ?' , 'is_complete': 'NPL hoàn tất trong ĐH ?' ,
                                         'po_status': 'Trạng thái ĐH' , 'product_code': 'Mã NPL', 'supplier_code' : 'Mã NCC' , 'supplier_name' : 'Tên NCC' , 'purchase_order_quantity' : 'Số lượng NPL đặt' , 'quantity_exchange' : 'Số lượng NPL đặt (đã chuyển đổi)' , 
                                        'quantity_arrived': 'Số lượng nhập kho', 'quantity_not_arrived': 'Số lượng chưa nhập kho' , 'time_create' : 'Thời gian tạo ĐH',
                                        'estimated_delivery' : 'Thời gian giao dự kiến' , 'actual_delivery_time' : 'Thời gian giao thực tế' , 'raw_material_status' : 'Tình trạng NPL'}, inplace = True)
dfPurchaseOrderDetailDA.head(5)
```
